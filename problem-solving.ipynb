        for(int i = 0; i < n-k-1; i++) {
            if(A[ i ] > A[ i+1] ) {
                // here swapping of positions is being done.
                temp = A[ i ];
                A[ i ] = A[ i+1 ];
                A[ i + 1] = temp;
            }
        }
    }
}

```python
import collections;

def bfsLevelFind(graph, level, lvlDict):
    queue = collections.deque([1]);
    res = 0;
    vis = [1];
    while queue:
        vertex = queue.popleft();
        vis.append(vertex);
        for i in graph[vertex]:
            if i not in vis:
                queue.append(i);
                vis.append(i)
                if(lvlDict[vertex]+1>level):
                    return res;
                elif(lvlDict[vertex]+1==level):
                    lvlDict[i] = lvlDict[vertex]+1;
                    res += 1;
                else:
                    lvlDict[i] = lvlDict[vertex]+1;
    return res;


n = int(input());
con = {};
lvlDict = {1: 1};
for i in range(n):
    con[i+1] = [];

for i in range(n-1):
    inp = list(map(int, input().split()));
    con[inp[0]].append(inp[1]);
    con[inp[1]].append(inp[0]);
# print(con)
level = int(input());
res = bfsLevelFind(con, level, lvlDict);
print(res);
```

### Using Kruskal

```python
v,e = [int(i) for i in input().split()]; 
graph = [];                 # Reading input from STDIN
for i in range(e):
    graph.append([int(j) for j in input().split()]);

parent = [0]*(v+1);
rank = [0]*(v+1);
for i in range(v):
    parent[i] = i;

def find(parent, x):
    if(parent[x]==x):
        return x;
    return find(parent, parent[x]);

def union(parent,x,y):
    if rank[x] < rank[y]:
        parent[x] = y;
    elif rank[x] > rank[y]:
        parent[y] = x
    else:
        parent[y] = x
        rank[x] += 1


def kruskal(graph, parent, ver):
    graph = sorted(graph, key=lambda x : x[2]);
    e_i=0;
    cnt=0;
    res_tre = [];
    min_dis = 0;
    while(cnt<ver-1):
        u,v,w = graph[e_i];
        e_i=e_i+1;
        x = find(parent,u);
        y = find(parent,v);
        if x!=y:
            cnt+=1;
            min_dis += w;
            res_tre.append([u,v,w]);
            union(parent,x,y);
    res = [res_tre,min_dis];
    return res;
print(kruskal(graph,parent,v)[1]);
```

### Using Prim's

```python
import sys;
v,e = [int(i) for i in input().split()]; 
graph = [];   
adj = [[0]*(v+1) for _ in range(v+1)];
for i in range(e):
    graph.append([int(j) for j in input().split()]);
for i in range(e):
    m,n,w = graph[i];
    adj[m][n] = w;
    adj[n][m] = w;
    # print(adj[m][n], adj[m][n], adj)
# print(graph)
# print(adj)
def prims(adj, ver):
    e_i = 0;
    selected = [False]*ver;
    selected[0] = True;
    res =0;
    while(e_i<ver-1):
        minimum = sys.maxsize;
        x = 0;
        y = 0;
        for i in range(ver):
            if selected[i]:
                for j in range(ver):
                    if (not selected[j] 
                    and adj[i+1][j+1]):
                        if minimum>adj[i+1][j+1]:
                            minimum = adj[i+1][j+1];
                            x = i;
                            y = j;
        # print(str(x) + "-" + str(y) + ":" + str(adj[x][y]))
        if(minimum!=sys.maxsize):
            res += minimum;
        selected[y]=True;
        e_i+=1;
    return res;
print(prims(adj,v));
```
